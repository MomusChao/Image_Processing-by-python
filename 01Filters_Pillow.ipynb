# Filters 

The current version of the library, Pillow (PIL Fork) 5.2.x, provides the following set of predefined image enhancement filters:

* BLUR
* CONTOUR
* DETAIL
* EDGE_ENHANCE
* EDGE_ENHANCE_MORE
* EMBOSS
* FIND_EDGES
* SHARPEN
* SMOOTH
* SMOOTH_MORE

# Load Lena image

```python
from PIL import Image
from PIL import ImageFilter
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
img = Image.open("lena_color.gif")
```

```python
plt.imshow(img)
```

# Convert to RGB image

```python
img2 = img.convert("RGB")
```

# Filter: BLUR

```python
imgfilted_b = img2.filter(ImageFilter.BLUR)
```

```python
plt.imshow(imgfilted_b)
```

# Filter: CONTOUR

It's kind of a profile filter

```python
imgfilted_c = img2.filter(ImageFilter.CONTOUR)
```

```python
plt.imshow(imgfilted_c)
```

# Filter: DETAIL

```python
imgfilted_DETAIL = img2.filter(ImageFilter.DETAIL)
```

```python
plt.imshow(imgfilted_DETAIL)
```

# Filter: EDGE_ENHANCE

```python
imgfilted_ee = img2.filter(ImageFilter.EDGE_ENHANCE)
```

```python
plt.imshow(imgfilted_ee)
```

# Filter: EDGE_ENHANCE_MORE

```python
imgfilted_ee_m = img2.filter(ImageFilter.EDGE_ENHANCE_MORE)
```

```python
plt.imshow(imgfilted_ee_m)
```

# Filter: EMBOSS

```python
imgfilted_em = img2.filter(ImageFilter.EMBOSS)
```

```python
plt.imshow(imgfilted_em)
```

# Filter: FIND_EDGES

It could offer edge information.

```python
imgfilted_fe = img2.filter(ImageFilter.FIND_EDGES)
```

```python
plt.imshow(imgfilted_fe)
```

# Filter: SHARPEN

```python
imgfilted_sh = img2.filter(ImageFilter.SHARPEN)
```

```python
plt.imshow(imgfilted_sh)
```

# Filter: SMOOTH

```python
imgfilted_sm = img2.filter(ImageFilter.SMOOTH)
```

```python
plt.imshow(imgfilted_sm)
```

# Filter: SMOOTH_MORE

```python
imgfilted_sm_m = img2.filter(ImageFilter.SMOOTH_MORE)
```

```python
plt.imshow(imgfilted_sm_m)
```
