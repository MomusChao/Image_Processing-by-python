# 1.Opencv python

gif file: use imageio module OR convert to 'jpg' or 'png'

```python
# convert to 'png'
from PIL import Image                   
 
im = Image.open("lena_color.gif")    
im.show()
im.save("lena_color.png","png") 
```

Read file

```python
import cv2
```

```python
img = cv2.imread('lena_color.png')
```

```python
type(img)
```

```python
img.shape
```

```python
# it would be error because image is BGR format
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
# change BGR to RGB
img_rgb = img[:,:,::-1]
# img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

grayscale

```python
# IMREAD_GRAYSCALE: 灰階
img_gray = cv2.imread('lena_color.png', cv2.IMREAD_GRAYSCALE)
```

```python
# setting: colormap 
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(img_gray)
```

```python
# change setting
plt.imshow(img_gray, cmap = 'gray')
plt.show()
```

Read RGB with transparency

```python
# 讀取圖片中所有的 channels，包含透明度的 channel。
img_UNCHANGED = cv2.imread('lena_color.png', cv2.IMREAD_UNCHANGED)
```

```python
# change BGR to RGB
img_UNCHANGED_rgb = img_UNCHANGED[:,:,::-1]
# img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(img_UNCHANGED_rgb)
```

Read RGB without transparency

```python
# 此為預設值，這種格式會讀取 RGB 三個 channels 的彩色圖片，而忽略透明度的 channel
img_IMREAD_COLOR = cv2.imread('lena_color.png', cv2.IMREAD_COLOR)
```

```python
# change BGR to RGB
# img_UNCHANGED_rgb = img_UNCHANGED[:,:,::-1]
img_IMREAD_COLOR_rgb = cv2.cvtColor(img_IMREAD_COLOR, cv2.COLOR_BGR2RGB)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(img_IMREAD_COLOR_rgb)
```
